```python
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
target_train = pd.read_csv('data/target_train.csv')

lom_train = pd.read_csv('data/lom_train.csv')
plavki_train = pd.read_csv('data/plavki_train.csv')

chugun_train = pd.read_csv('data/chugun_train.csv')
sip_train = pd.read_csv('data/sip_train.csv')
gas_train = pd.read_csv('data/gas_train.csv')
produv_train = pd.read_csv('data/produv_train.csv')
chronom_train = pd.read_csv('data/chronom_train.csv')
```

```python
def get_seconds(value, format):
    time_to_proc = datetime.strptime(value, format)
    diff = time_to_proc - datetime(2021, 1, 1)
    secs = round(diff.total_seconds(), 1)
    return int(secs)
```

```python
chugun_train['timed'] = chugun_train['DATA_ZAMERA'].apply(lambda x : get_seconds(x, "%Y-%m-%d %H:%M:%S"))
```

```python
sip_train['timed'] = sip_train['DAT_OTD'].apply(lambda x : get_seconds(x, "%Y-%m-%d %H:%M:%S"))
```

```python
produv_train['timed'] = produv_train['SEC'].apply(lambda x : get_seconds(x, "%Y-%m-%d %H:%M:%S"))
```

```python
gas_train['timed'] = gas_train['Time'].apply(lambda x : get_seconds(x, "%Y-%m-%d %H:%M:%S.%f"))
```

```python
chronom_train['timed'] = chronom_train.dropna()['VR_NACH'].apply(lambda x : get_seconds(x, "%Y-%m-%d %H:%M:%S"))
chronom_train = chronom_train.dropna()
```

```python
merged_data = pd.merge_asof(
    gas_train.sort_values(by='timed'), produv_train.sort_values(by='timed'),
    on='timed', direction='backward'
)
```

```python
merged_data_2 = pd.merge_asof(
    merged_data.sort_values(by='timed'), chugun_train.sort_values(by='timed'),
    on='timed', direction='backward'
)
```

```python
merged_data_3 = pd.merge_asof(
    merged_data_2.sort_values(by='timed'), sip_train.sort_values(by='timed'),
    on='timed', direction='backward'
)
```

```python
chronom_train['timed'] = chronom_train['timed'].astype('int64')
```

```python
merged_data_4 = pd.merge_asof(
    merged_data_3.sort_values(by='timed'), chronom_train.sort_values(by='timed'),
    on='timed', direction='backward'
)
```

```python
merged_data_4 = merged_data_4[[
                 'Time',
                 'V_x',
                 'T_x',
                 'O2_x',
                 'N2',
                 'H2',
                 'CO2',
                 'CO',
                 'AR',
                 'O2_pressure',
                 'timed',
                 'SEC',
                 'RAS',
                 'POL',
                 'VES',
                 'T_y',
                 'SI',
                 'MN',
                 'S',
                 'P',
                 'CR',
                 'NI',
                 'CU',
                 'V_y',
                 'TI',
                 'DATA_ZAMERA',
                 'VDSYP',
                 'NMSYP',
                 'VSSYP',
                 'DAT_OTD',
                 'NPLV',
                 'TYPE_OPER',
                 'NOP',
                 'VR_NACH',
                 'VR_KON',
                 'O2_y']]
```

```python
merged_data_4.to_csv('merged_data.csv', index=False)
```
